```python
import pandas as pd
```

```python
df = pd.read_csv("../Resources/acclaimed_albums_genres.csv")

display(df.head())
print(f"{df.info()}")
df.columns
```

```python
df = df.convert_dtypes(str)
df.info()
```

```python
df_genres = df[['Genres']]
display(df_genres.value_counts())
display(df_genres.info())
```

```python
import warnings
# Suppress FutureWarning messages
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
# split out the values that have more than one genre
df_genres_split = df_genres.copy()
df_genres_split[['Main Genre', 'Sec Genre', '3rd Genre', '4th Genre']] = df_genres_split['Genres'].str.split(',',expand=True)
display(df_genres_split.head(20))
df_genres_split.columns
```

```python
# display(f"{count(df_genres['Secondary Genres'].notna())}")
display(f"{df_genres_split['Main Genre'].count()}")
display(f"{df_genres_split['Sec Genre'].count()}")
display(f"{df_genres_split['3rd Genre'].count()}")
display(f"{df_genres_split['4th Genre'].count()}")

display(df_genres_split['Main Genre'].value_counts())
display(df_genres.info())
```

```python
one_hot = pd.get_dummies(df_genres_split['Main Genre'].apply(pd.Series).stack()).groupby(level=0).sum()
df_one_hot = df_genres.join(one_hot)
df_one_hot.drop(columns=['Genres'], inplace=True)

display(df_one_hot.head())
display(df_one_hot.info())
```

```python
# remove least used Categories, finding smallest occurrences in the dataset
genre_counts = {}
display(len(df_one_hot[df_one_hot["Rock"] == 1]))

genre_list = df_one_hot.columns
# ['Rock', 'Pop', 'Baroque Pop', 'Acid House', 'Acid Jazz', 'Acid Rock', 'Acoustic Rock']
print(list)

for genre in genre_list:
    num = len(df_one_hot[df_one_hot[genre] == 1])
    genre_counts[genre] = num

display(genre_counts)
```

```python
sorted_Dict = dict(sorted(
        genre_counts.items(), reverse=True, key= lambda x : x[1] ) )
# Getting the keys from the sorted dictionary
# sorted_keys = sorted_Dict.keys()

# Converting the keys to a list
keys_list = list(sorted_Dict)

# display(type(sorted_keys))
display(keys_list[:20])
```

```python
df_top_genres = df_one_hot[['Alternative Rock',
 'Indie Rock', 'Pop Rock', 'Art Pop', 'Art Rock', 'Blues Rock', 'Hard Rock', 
 'New Wave', 'Folk Rock', 'Indie Pop', 'Hard Bop',
 'Alt-Country', 'Contemporary Folk', 'Post-Punk', 'Contemporary R&B',
 'Singer/Songwriter', 'Chamber Pop', 'Conscious Hip Hop', 'Avant-Garde Jazz',
 'Britpop']]
    
    # old list
# ['Rock', 'Pop', 'Pop Rock', 'Alternative Rock', 'R&B', 'Hip Hop', 'Hard Rock',
#  'Singer/Songwriter', 'Electronic', 'Soul', 'Metal', 'Contemporary R&B', 'Dance',
#  'Pop Soul', 'Hardcore Hip Hop', 'Alternative Metal', 'Blues Rock', 'New Wave',
#  'Folk Rock', 'Progressive Rock']]

display(df_top_genres.head())
display(df_top_genres.info())
```

```python
# Add year column from Acclaimed Dataframe
df_top_genres['Year']= df['Year']

display(df_top_genres.head())
display(df_top_genres.info())
display(df_top_genres.columns)

# df_combined = pd.concat([df, df_top_genres], axis=1)
# display(df_combined.head())
# display(df_combined.info())
# display(df_combined.columns)

# next: group by year, sum
```

```python
# Group by year and get sums of each column related to genre:
genres_by_year_df = df_top_genres.groupby('Year').sum(numeric_only=True)
genres_by_year_df.info()
genres_by_year_df
```

```python
# Create Totals column, insert sum the total entries for each year (each row):

row_sums = genres_by_year_df.sum(axis=1)
genres_by_year_df['Total'] = row_sums
first_column = genres_by_year_df.pop('Total') 
genres_by_year_df.insert(0, 'Total', first_column) 
display(genres_by_year_df)
display(genres_by_year_df.info())
# display(list(row_sums))
```

```python
df_drops = genres_by_year_df[genres_by_year_df['Total'] > 5]
df_drops.info()
```

```python
# Divide each element by the Total row sum and multiply by 100 to get percentages
# df2_percentage = df2.div(row_sums, axis=0) * 100

df_percent = genres_by_year_df.copy()
df_percent = round(df_percent.div(row_sums, axis=0), 4)
# display(df_percent.head())

# replace NaN with 0
df_percent.fillna(0, inplace=True)

# drop total column
df_percent.drop(columns='Total', inplace=True)
# rename each column to have %
df_percent_final = df_percent.add_suffix(' %')

display(df_percent_final)
```
